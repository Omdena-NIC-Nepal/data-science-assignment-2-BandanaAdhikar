 ## 1.Handle missing values


```python

# Remove rows where Name is 'invalid' or NaN
 #Try converting #'Age' to numeric, handling errors as NaN
```

```python

# Remove rows where Name is 'invalid' or NaN
 #Try converting #'Age' to numeric, handling errors as NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Now you can safely calculate the mean and use it for filling
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df) 

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')  # Convert 'Salary' to numeric, coercing errors to NaN
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df)
```

```python
# Fill missing categorical values with the mode of their respective columns
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])
df['Email'] = df['Email'].fillna(df['Email'].mode()[0])
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
df['DateOfJoining'] = df['DateOfJoining'].fillna(df['DateOfJoining'].mode()[0])
```

```python
# Display the dataset after handling missing values
print("\nDataset after handling missing values:")
print(df.head())
```

## 2.Handling Invalid Values


```python

# Replace invalid 'Age' and 'Salary' entries with NaN, then fill with mean
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
# Display the dataset after handling invalid values
print("\nDataset after handling invalid values:")
print(df.head())
```

## 3. Duplicates Removal


```python

# Check for duplicates
duplicates = df.duplicated()
print("\nNumber of duplicate rows:", duplicates.sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Display the dataset after removing duplicates
print("\nDataset after removing duplicates:")
print(df.head())
```

## 4. Consistent Formatting



```python
# 4. Consistent Formatting
# Ensure consistent capitalization for 'Name' and 'Country'
df['Name'] = df['Name'].str.title()
df['Country'] = df['Country'].str.title()
```

```python
# Ensure 'Email' is lowercase
df['Email'] = df['Email'].str.lower()
# Display the final cleaned dataset
print("\nFinal Cleaned Dataset:")
print(df.head())
```

```python
# Save the cleaned dataset
cleaned_file_path = ('/Users/bandanaadhikari/Desktop/classassignment/cleanedmessed_dataset.csv')
df.to_csv(cleaned_file_path, index=False)
```

```python
# Check for non-numeric values in the 'Age' column
non_numeric_ages = df[~df['Age'].astype(str).str.replace('.', '', 1).str.isnumeric()]
print("\nNon-numeric values in the 'Age' column:")
print(non_numeric_ages)

```

```python
# Replace 'invalid' entries
df['Name'].replace('invalid', 'Unknown', inplace=True)
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# Replace missing values
df['Name'].fillna('Unknown', inplace=True)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert age to numeric, forcing errors to NaN
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Email'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')  # Convert salary to numeric, forcing errors to NaN
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['DateOfJoining'].fillna(df['DateOfJoining'].mode()[0], inplace=True)
```

```python
# Convert infinite values to NaN
df.replace([float('inf'), float('-inf')], float('nan'), inplace=True)
```

```python
# Plot Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
 #Plot Salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True, bins=30)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot count of individuals per Country
plt.figure(figsize=(12, 8))
sns.countplot(y='Country', data=df, order=df['Country'].value_counts().index)
plt.title('Count of Individuals per Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```
